```python
# importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading the datasets
data = pd.read_csv("data.csv")
data.head()
```

# Univariate Analysis

```python
# visualizing the the distribution of Age
plt.figure(figsize=(15,20))
plt.subplot(4,3,1)
sns.histplot(data["Age"],kde=True)
plt.title("Age distribution")

plt.subplot(4,3,2)
sns.histplot(data["Weight (kg)"],kde=True)
plt.title("Weight Distribution")

plt.subplot(4,3,3)
sns.histplot(data["Height (m)"],kde=True)
plt.title("Height Distribution")

```

# Bivariate Analysis

```python

sns.scatterplot(x="Age",y="Weight (kg)",data=data,hue="Gender")
plt.title("Age vs Weight")
```

The graph shows the distribution of weight spread out from 20 years to 60 years 

males(marked in blue) appear to have wide range of weight reaching upto to 120kg 
females(marked in orange) generally have lower weights compared to males clusttering between 40 and 80

```python

sns.barplot(x="Workout_Type", y="Calories_Burned", data=data)
plt.title("Calories Burned by Workout Type")
```

All the 4 workouts have similar Average number of calories_Burned ,on average these workout types maybe comparable interms of calories expenditure
